```python
prefix查询和过滤器,和term查询类似：prefix查询和过滤器允许你根据指定的前缀搜索词条，(对于event-index的索引,匹配所有以liber开头的文档)
```

```python
curl -H "Content-Type:application/json" -XPOST 'localhost:9200/event-index/event/_search?pretty' -d '{
    "query": {
        "prefix": {
            "title": "liber"
        }
    }
}'
```

```python
curl -H "Content-Type:application/json" -XPOST 'localhost:9200/event-index/event/_search?pretty' -d '{
    "query": {
        "bool": {
            "must": {
                "match_all": {}
            },
            "filter": {
                "prefix": {
                    "title": "liber"
                }
            }
        }
    }
}'
```
